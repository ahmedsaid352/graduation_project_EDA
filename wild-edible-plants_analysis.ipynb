```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('datasetinfo2.xlsx')
```

```python
df.head()
```

```python
df['number of images']=df['number of images'].apply(lambda x: x.split(' ')[0])
```

```python
df.head()
```

```python
df['number of images']=df['number of images'].astype(int)
```

```python
df.info()
```

```python
plt.figure(figsize=(15, 8))
plt.barh(df['plant name'],df['number of images']);
plt.title('PlantVillage Dataset',fontsize=20)
plt.ylabel('Plant Name',fontsize=18)
plt.xlabel('Number of images',fontsize=18)
# plt.xlim([390, 510])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('bar-chart2.jpg')
plt.show()
```

```python
df.to_csv('dataset2V2.csv',index=False)
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='number of images',data=df);
plt.xlabel('')
plt.title('Distribution of number of images',fontsize=14)
plt.xticks(fontsize=14)
plt.show()
```
